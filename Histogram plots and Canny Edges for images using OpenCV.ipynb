```python
# importing dependencies
import cv2
import urllib.request
```

```python
#Set the url and filename

url = "https://images.unsplash.com/photo-1577046084510-9bd224305e7e?ixlib=rb-1.2.1&ixid=eyJhcHBfaWQiOjEyMDd9&auto=format&fit=crop&w=500&q=60"
filename="test_image.jpg"

urllib.request.urlretrieve(url, filename)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread(filename)

plt.imshow(image)
```

```python
img_corrected = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(img_corrected)
```

```python
plt.axis("off") #remove axes ticks
plt.imshow(img_corrected)
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = 12, 17

plt.axis("off") #remove axes ticks
plt.imshow(img_corrected)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap = 'gray')
plt.axis("off") #remove axes ticks
plt.title('Grayscale Image')
```

#### Applying Canny Edge detector to this image with varying threshold levels

```python
rcParams['figure.figsize'] = 10, 12


edges = cv2.Canny(img_corrected, 
                  threshold1=100, 
                  threshold2=200)

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

```python
rcParams['figure.figsize'] = 10, 12

edges = cv2.Canny(img_corrected, 
                  threshold1=1,   ## try different values here
                  threshold2=200) ## try different values here

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

```python
rcParams['figure.figsize'] = 10, 12

edges = cv2.Canny(img_corrected, 
                  threshold1=100,   ## try different values here
                  threshold2=500) ## try different values here

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

### Using Histograms

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

rcParams['figure.figsize'] = 8,4

plt.hist(gray_image.ravel(),256,[0,256])
plt.title('Histogram of Grayscale image.jpg')
plt.show()
```

### RGB Histogram

```python
rcParams['figure.figsize'] = 8, 4

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
